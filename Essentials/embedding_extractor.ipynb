# Embeddings

```python
import pandas as pd
import pickle as pk
from tqdm import tqdm
from sentence_transformers import SentenceTransformer
```

```python
data = pd.read_csv("./Data/processed-synth-auditor-stream.csv")
```

```python
model = SentenceTransformer('all-mpnet-base-v2')
```

```python
data['news'] = data['news'].astype('str')
```

```python
classes = data['class']
```

```python
classes[1]
```

```python
import time
import pickle as pk

stamp = "-".join(str(time.time()).split("."))
file_path = f"./Data/_143664_v02_embeddings/sentence-transformer-vector-embedding-{stamp}.pkl"

with open(file_path, 'wb') as file:
  pk.dump(([1,2,3,4,5], 0), file) # ([embeddings], class_label)
```

```python
import time
import pickle as pk
import datetime 

log_file = open('./log.txt', 'a')

embeddings = []
idx = 0
for text in tqdm(data['news'], desc="Generating embeddings"):
    stamp = "-".join(str(time.time()).split("."))
    emb = model.encode(text)
    with open(f"./Data/_143664_v02_embeddings/sentence-transformer-vector-embedding-{stamp}.pkl", 'wb') as file:
        pk.dump((emb, classes[idx]), file)
    log_data = f"file content - data[{idx}] has been pickled at {datetime.datetime.now()} into sentence-transformer-vector-embedding-{stamp}.pkl successfully\n"
    print(log_data, end="")
    log_file.write(log_data)
    embeddings.append(emb)
    idx += 1
```
